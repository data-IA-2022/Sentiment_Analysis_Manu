### Evaluation d'indépendance entre variables qualitatives à l'aide du test du Chi².

```python
from scipy.stats import chi2_contingency

categorical_vars = df.select_dtypes(exclude = 'number').columns

print("-----------------------")

for vr in categorical_vars:
    for var in categorical_vars:
        if vr != var:
            crosstab = pd.crosstab(df[var], df[vr])
            
            # calcul du test du Chi² et affichage du résultat
            chi2, p, dof, expected = chi2_contingency(crosstab)

            print(f"P-value {vr} vs {var} = {np.round(p,3)}")
            
            print("-----------------------")
```


- Les Chi² sont tous inpérieur au seuil fixé à 0.05.
- On peut conclure que les variables ne sont pas significativement associées les unes aux autres, les données ne s'influancent pas mutuelement.


# Mise en forme du dataset


- Il y a très peu d'entrées dans lesquelles des champs ne sont pas renseignés (393, soit 0.3% de l'ensemble des données). Par conséquent, il est raisonnable de supprimer ces entrées.
- Les données quantitatives présentent des ordres de grandeur très disparates, ce qui peut affecter la comparaison des variables. Afin de corriger cela, il est proposé de normaliser les données en utilisant des méthodes telles que la normalisation min-max.


```python
#Suppression des entrées avec 'Arrival Delay in Minutes' à None
df_quant_dts = df_quant[df['Arrival Delay in Minutes'].notnull()]
df_qualit_dts = df_qualitatives[df['Arrival Delay in Minutes'].notnull()]
df_ordinal_dts = df_ordinal[df['Arrival Delay in Minutes'].notnull()]
df_qualit_dts.shape
```

```python
df_dts = df_quant_dts.join(df_qualit_dts)
df_dts = df_dts.join(df_ordinal_dts)
df_dts.shape
```

```python
df_dts.columns
```


# Processus de machine learning
## Sentiment Analysis - Airline Passenger Satisfaction

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import svm
from sklearn import datasets
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import NearestNeighbors
```


- Extraction de la colonne label (cible ou variable dépendante)
- Division de la base de données en deux parties : une partie pour l'entraînement (train) et une autre pour le test (test).


```python

y=df_dts['Satisfaction']
X=df_dts.drop('Satisfaction', axis=1)
#Split de la base
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, stratify=y)
```

```python
X_train
```

```python
y_train.shape
```


- Partage des features entre numériques et catégorielles.


```python
cat_cols=X_train.select_dtypes(exclude = 'number').columns
num_cols=X_train.select_dtypes(include = ['int', 'float']).columns
```

```python
cat_cols
```

```python
num_cols
```


### Construction du pipeline pour la génération d'un modèle de classification.
#### Prétraitement
    - Normalisation et encodage des features

```python
num_pipeline = make_pipeline(MinMaxScaler())
```

```python
#num_pipeline
```

```python
cat_pipeline = make_pipeline(OneHotEncoder())
```

```python
#cat_pipeline
```

```python
#Transformer
mct=make_column_transformer((cat_pipeline, cat_cols),(num_pipeline,num_cols))
```

```python
#mct
```


 #### Déclaration et ajout de l'algorithme de classification.
    

```python
full_pipeline = make_pipeline(mct, KNeighborsClassifier(n_neighbors=3))
```

```python
full_pipeline
```


 #### Entraînement et génération du modèle de classification.
    

```python
full_pipeline.fit(X_train,y_train)
```

    
#### Évaluation du modèle de classification sur un jeu de test indépendant du jeu d'entraînement.
    

```python
full_pipeline.score(X_test,y_test)
```


### Construction du pipeline de recherche d'un meilleur modèle, de son évaluation et de son interprétation.


```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

#supprimer les warnings
import warnings
warnings.filterwarnings("ignore")
```


#### Paramètres pour la méthode de validation croisée.
    

```python
cv = StratifiedKFold(n_splits=5) #Equivalent à KFold ou cv=None dans ce context, mais explicité
```


#### On récupère le preprocessing du pipeline suivant, auquel on ajoute une référence au modèle (model).


```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
full_pipeline_grid = Pipeline([('mct',mct),
                               ('model', None)])
```

```python
full_pipeline_grid
```

    
#### Déclaration des algorithmes de classification et des hyper-paramètres que l'on souhaite évaluer afin de faire émerger la meilleure combinaison.
    

```python
param_grid = [
     {
        'model': [RandomForestClassifier ()],
        'model__max_depth': [80, 100],
        'model__min_samples_split': [8, 10, 12],
        'model__n_estimators': [100, 200, 300]
    },
    {
        'model': [KNeighborsClassifier()],
        'model__n_neighbors': [3, 5, 7],
    },
    {
        'model': [GaussianNB()],
        'model__var_smoothing': [1e-9, 1e-7, 1e-5],
    },
    {
        'model': [SVC()],
        'model__kernel': ['linear', 'rbf'],
        'model__C': [0.1, 1],
    }
]
```

    
#### Construction du GrisSearch.
    

```python
grid = GridSearchCV(full_pipeline_grid, param_grid=param_grid, scoring='accuracy', cv=cv, verbose=2)
```


#### Entraînement.


```python
grid.fit(X_train, y_train)
```

    
#### Prédiction sur un jeu de données indépendant de l'entraînement.
    

```python
y_pred= grid.predict(X_test)
```

    
#### Évaluation de la performance de l'algorithme associé aux meilleurs hyperparamètres.
    

```python
from sklearn.metrics import accuracy_score
```

```python
best_model = grid.best_estimator_
best_model.fit(X_train, y_train)
```

```python
{'accuracy':{'pred': np.round(accuracy_score(y_test, y_pred),3), 'test':  np.round(grid.best_score_, 3)}, 'best_params_':grid.best_params_}
```
